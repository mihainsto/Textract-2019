```python
df = pd.read_csv('/kaggle/input/textract-2019/textract_train.csv')
```

```python
from googletrans import Translator
translator = Translator()
```

```python
def augment_translate(article):
    a = translator.translate(article, src='en', dest='fr')
    a = translator.translate(a.text, src='fr', dest='de')
    b = translator.translate(a.text, src='de', dest='en')
    result_text = b.text
    return result_text
```

```python
errCount = 0
with open('augmented_data0','w') as f:
    for i in tqdm(range(len(df['Contents']))):
        sentence = df['Contents'][i]
        if df['Label'][i] != 0:
            continue
        try:
            augmented_sentence = augment_translate(sentence)
        except:
            errCount += 1
        f.write(augmented_sentence)
        f.write('\01012340120100')
print(errCount)
```
