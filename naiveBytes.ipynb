```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from collections import Counter
import nltk
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score
```

```python
df = pd.read_csv('/kaggle/input/textract-2019/textract_train.csv')
```

```python
df['Contents']
```

```python
tfidf = TfidfVectorizer(ngram_range=(1,2), max_df= 0.85, min_df= 0.01)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Contents'], df['Label'], test_size = 0.2, random_state=42, stratify = df['Label'])
```

```python
X_train_vec = tfidf.fit_transform(X_train)
```

```python
X_test_vec = tfidf.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb_body = MultinomialNB()
nb_body.fit(X_train_vec, y_train)
pred = nb_body.predict(X_test_vec)
print ('Accuracy score {:.4}%'.format(accuracy_score(y_test, pred)*100))
```
